```python
# Given temperature data for Tribhuwan International Airport, Kathmandu
temperatures = [
    8.2, 17.4, 14.1, 7.9, 18.0, 13.5, 9.0, 17.8, 13.0, 8.5,
    16.5, 12.9, 7.7, 17.2, 13.3, 8.4, 16.7, 14.0, 9.5, 18.3, 
    13.4, 8.1, 17.9, 14.2, 7.6, 17.0, 12.8, 8.0, 16.8, 13.7, 
    7.8, 17.5, 13.6, 8.7, 17.1, 13.8, 9.2, 18.1, 13.9, 8.3, 
    16.4, 12.7, 8.9, 18.2, 13.1, 7.8, 16.6, 12.5
]

# Task 1: Classify temperatures into categories

# Initialize empty lists for classifications
cold = []        # Temperatures below 10°C
mild = []        # Temperatures between 10°C and 15°C
comfortable = [] # Temperatures between 15°C and 20°C

# Classify each temperature
for temp in temperatures:
    if temp < 10:
        cold.append(temp)
    elif 10 <= temp < 15:
        mild.append(temp)
    elif 15 <= temp < 20:
        comfortable.append(temp)

# Output the categorized lists
print("Cold temperatures (<10°C):", cold)
print("Mild temperatures (10°C to <15°C):", mild)
print("Comfortable temperatures (15°C to <20°C):", comfortable)
```

```python
import matplotlib.pyplot as plt  # Import matplotlib for plotting

# Given temperatures list
temperatures = [
    8.2, 17.4, 14.1, 7.9, 18.0, 13.5, 9.0, 17.8, 13.0, 8.5,
    16.5, 12.9, 7.7, 17.2, 13.3, 8.4, 16.7, 14.0, 9.5, 18.3,
    13.4, 8.1, 17.9, 14.2, 7.6, 17.0, 12.8, 8.0, 16.8, 13.7,
    7.8, 17.5, 13.6, 8.7, 17.1, 13.8, 9.2, 18.1, 13.9, 8.3,
    16.4, 12.7, 8.9, 18.2, 13.1, 7.8, 16.6, 12.5
]

# 1. Create empty lists for night, evening, and day temperatures
night_temperatures = []
evening_temperatures = []
day_temperatures = []

# 2. Iterate over the list and group temperatures by time of day
for i in range(len(temperatures)):
    if i % 3 == 0:  # Night readings (00-08)
        night_temperatures.append(temperatures[i])
    elif i % 3 == 1:  # Evening readings (08-16)
        evening_temperatures.append(temperatures[i])
    elif i % 3 == 2:  # Day readings (16-24)
        day_temperatures.append(temperatures[i])

# 3. Calculate and print the average day-time temperature
average_day_temp = sum(day_temperatures) / len(day_temperatures)
print(f"Average day-time temperature: {average_day_temp:.2f} °C")

# 4. Optional: Plot "day vs. temperature" using matplotlib
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(day_temperatures) + 1), day_temperatures, marker='o', label="Day-time Temperatures")
plt.title("Day vs. Temperature (°C)")
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Recursive function to sum all numbers in a nested list
def sum_nested_list(nested_list):
    total_sum = 0  # Initialize the sum
    for item in nested_list:
        if isinstance(item, list):  # Check if the item is a list
            total_sum += sum_nested_list(item)  # Recursively call the function
        else:
            total_sum += item  # Add the number to the total sum
    return total_sum

# Sample nested list for testing
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]

# Call the function and print the result
result = sum_nested_list(nested_list)
print(f"Total sum of nested list: {result}")
```

```python
# Recursive function to generate all permutations of a string
def generate_permutations(s):
    # Base case: if the string has only one character, return it as a list
    if len(s) == 1:
        return [s]

    # Recursive case
    permutations = []
    for i, char in enumerate(s):
        # Exclude the current character and generate permutations of the remaining string
        remaining = s[:i] + s[i+1:]
        for perm in generate_permutations(remaining):
            # Add the current character to each permutation of the remaining string
            permutations.append(char + perm)
    
    # Return unique permutations by converting to a set and back to a list
    return list(set(permutations))

# Test the function with sample inputs
print("Permutations of 'abc':", generate_permutations("abc"))
print("Permutations of 'aab':", generate_permutations("aab"))
```

```python
def calculate_directory_size(directory):
    """
    Recursively calculates the total size of a directory, including all files and nested subdirectories.

    Args:
    - directory (dict): The directory structure as a dictionary.

    Returns:
    - int: Total size of the directory in KB.
    """
    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict):
            # If the value is a dictionary, recursively calculate its size
            total_size += calculate_directory_size(value)
        else:
            # If the value is an integer, it represents a file size
            total_size += value
    return total_size

# Example directory structure
file_system = {
    "file1.txt": 100,
    "file2.txt": 200,
    "subdir1": {
        "file3.txt": 300,
        "file4.txt": 400,
        "subdir2": {
            "file5.txt": 500
        }
    },
    "subdir3": {
        "file6.txt": 600
    }
}

# Calculate the total size of the root directory
total_size = calculate_directory_size(file_system)
print(f"Total size of the directory: {total_size} KB")
```

```python
def min_coins(coins, amount):
    """
    Calculates the minimum number of coins needed to make up the given amount.

    Args:
        coins (list): List of coin denominations.
        amount (int): Target amount.

    Returns:
        int: Minimum number of coins needed or -1 if the amount cannot be formed.
    """
    # Initialize the DP array with a value greater than the maximum possible answer
    dp = [float('inf')] * (amount + 1)
    
    # Base case: 0 coins are needed to make amount 0
    dp[0] = 0
    
    # Fill the DP array
    for coin in coins:
        for x in range(coin, amount + 1):
            dp[x] = min(dp[x], dp[x - coin] + 1)
    
    # If the amount cannot be formed, return -1
    return dp[amount] if dp[amount] != float('inf') else -1
# Test case
coins = [1, 2, 5]
amount = 11

# Calculate the minimum number of coins
result = min_coins(coins, amount)

# Print the result
print(f"Minimum coins required to make {amount}: {result}")
```

```python
def longest_common_subsequence(s1, s2):
    """
    Calculates the length of the longest common subsequence (LCS) between two strings.

    Args:
        s1 (str): The first string.
        s2 (str): The second string.

    Returns:
        int: The length of the LCS.
    """
    # Lengths of the input strings
    m, n = len(s1), len(s2)
    
    # Create a DP table with dimensions (m+1) x (n+1) initialized to 0
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    # Fill the DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                # Characters match, add 1 to the diagonal value
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                # Characters do not match, take the max of the left and top values
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    
    # The bottom-right corner contains the length of the LCS
    return dp[m][n]
    # Test case
s1 = "abcde"
s2 = "ace"

# Calculate the LCS length
lcs_length = longest_common_subsequence(s1, s2)

# Print the result
print(f"The length of the LCS between '{s1}' and '{s2}' is: {lcs_length}")

```

```python
def knapsack(weights, values, capacity):
    """
    Solves the 0/1 knapsack problem using dynamic programming.

    Args:
        weights (list): A list of item weights.
        values (list): A list of item values.
        capacity (int): The maximum weight capacity of the knapsack.

    Returns:
        int: The maximum value that can be carried in the knapsack.
    """
    n = len(weights)
    
    # Create a DP table with (n+1) rows and (capacity+1) columns
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    
    # Fill the DP table
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                # If the item can be included, take the maximum of including it or excluding it
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                # If the item cannot be included, just carry forward the value without it
                dp[i][w] = dp[i - 1][w]
    
    # The maximum value is in the bottom-right corner of the DP table
    return dp[n][capacity]
# Test case
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7

# Calculate the maximum value
max_value = knapsack(weights, values, capacity)

# Print the result
print(f"The maximum value that can be carried is: {max_value}")
```
